```python
import spacy
nlp = spacy.load('en_core_web_sm')
from spacy import displacy
```

```python
with open('peterrabbit.txt') as rd:
    peter_rabbit=rd.read()
```

```python
print(peter_rabbit)
```

```python
doc=nlp(peter_rabbit)
```

```python
peter_rabbit_sent_list=list(doc.sents)
```

```python
for i in range(len(peter_rabbit_sent_list)):
    if(i%3==0):
        doc_i=nlp(str(peter_rabbit_sent_list[i]))
        for token in doc_i:
            print(f"{token.text:{10}}{token.pos_:{10}} {token.tag_:{10}}{str(spacy.explain(token.tag_))}")
    
```

```python
for token in list(doc.sents)[2]:
    print(f"{token.text:{10}}{token.pos_:{10}} {token.tag_:{10}}{str(spacy.explain(token.tag_))}")
    
```

```python
POS_counts=doc.count_by(spacy.attrs.POS)
```
